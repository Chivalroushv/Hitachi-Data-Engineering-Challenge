## Feature Engineering

```python
train[['station','source_no']]=train.source_name.str.split('$',expand=True)
train[['station','destination_no']]=train.destination_name.str.split('$',expand=True)
```

```python
train=train.drop(['source_name','destination_name','station'],axis=1)
train['source_no']=train.source_no.astype(int)
train['destination_no']=train.destination_no.astype(int)
```

```python
train['time']=pd.to_datetime(train.current_time).dt.time
test['time']=pd.to_datetime(test.current_time).dt.time
```

```python
# rows in train data where source and destination are same
train[train.source_no==train.destination_no]
```

```python
#rows in test data where source and destination are same
test[test.source_name==test.destination_name]
```

```python
#creating a new column zone and inserting random value
#this random value will replaced with morning, evening, afternoon and night in next cell
for i in train.index:
    train['zone']=1
for i in test.index:
    test['zone']=1
```

```python
#on the basis of time, splitting 24 hrs in morning, eve, afternoon as after and night
for i in test.index:
    if((test.time[i]>=datetime.time(0,0,0))&(test.time[i]<datetime.time(6,0,0))):
        test.zone[i]='night'
    if((test.time[i]>=datetime.time(6,0,0))&(test.time[i]<datetime.time(12,0,0))):
        test.zone[i]='morning'
    if((test.time[i]>=datetime.time(12,0,0))&(test.time[i]<datetime.time(18,0,0))):
        test.zone[i]='after'
    if((test.time[i]>=datetime.time(18,0,0))&(test.time[i]<=datetime.time(23,59,59))):
        test.zone[i]='eve'
for i in train.index:
    if((train.time[i]>=datetime.time(0,0,0))&(train.time[i]<datetime.time(6,0,0))):
        train.zone[i]='night'
    if((train.time[i]>=datetime.time(6,0,0))&(train.time[i]<datetime.time(12,0,0))):
        train.zone[i]='morning'
    if((train.time[i]>=datetime.time(12,0,0))&(train.time[i]<datetime.time(18,0,0))):
        train.zone[i]='after'
    if((train.time[i]>=datetime.time(18,0,0))&(train.time[i]<=datetime.time(23,59,59))):
        train.zone[i]='eve'
```

```python
#creating a new column "is_weekend" to check whether its a weekend or not
for i in train.index:
    if train.current_date[i]=='2016-09-05':
        train.is_weekend[i]=True
for i in test.index:
    if test.current_date[i]=='2016-10-10':
        test.is_weekend[i]=True
```

```python
train['zone'].replace(['night','morning','after','eve'],[0,2,1,3],inplace=True)
test['zone'].replace(['night','morning','after','eve'],[0,2,1,3],inplace=True)
```

```python
#preprocessing data with standard scaler
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

### Random Forest Clasifier

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
tscv=TimeSeriesSplit(n_splits=5)
rfc=RandomForestClassifier(n_estimators=50,max_depth=3,min_samples_split=0.1,random_state=11)
```

```python
train.columns
```

```python
from sklearn.model_selection import train_test_split
x=train[[ 'longitude_source', 'latitude_source','mean_halt_times_source','longitude_destination',
                'latitude_destination','mean_halt_times_destination', 'is_weekend','zone']]
x=scaler.fit_transform(x)
y=train[['target']]
x_train,x_test,y_train,y_test=train_test_split(x,y,stratify=y,test_size=0.2,random_state=1111)
rfc.fit(x_train,y_train)
print('test score:',f1_score(y_test,rfc.predict(x_test),average='weighted'))
print('train score:',f1_score(y_train,rfc.predict(x_train),average='weighted'))
```

### Prediction on test dataset

```python
x=test[[ 'longitude_source', 'latitude_source','mean_halt_times_source','longitude_destination',
                'latitude_destination','mean_halt_times_destination', 'is_weekend','zone']]
x=scaler.transform(x)
```

```python
y_pred=rfc.predict(x)
```
